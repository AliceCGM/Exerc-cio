```python
df = df.rename(columns={"ID": "Identificador",
                        "Pressão sanguíneaaaa": "Pressão Sanguínea",
                       "Ocupação": "Profissão",
                            "Categoria BMI": "Categoria IMC"})
df.head()
```

```python
#Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões?
média = df.groupby('Profissão')['Duração do sono'].mean()
mediana = df.groupby('Profissão')['Duração do sono'].median()
moda = df.groupby('Profissão')['Duração do sono'].agg(mode)
print(média, mediana, moda)
```

```python
#4. De acordo com os dados, advogar ou ser representante de vendas faz você dormir menos? (Use o método ‘isin’, considere a média)
profissoes_interesse = df[df['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])]
medias = profissoes_interesse.groupby('Profissão')['Duração do sono'].mean()
profissao_que_dorme_menos = medias.idxmin()

print(f"A profissão que dorme menos é: {profissao_que_dorme_menos}")
print(f"Médias de sono:\n{medias}")
```

```python
#8. Quem tem maior pressão sanguínea média, homens ou mulheres? (Considere a média)
def categorizar_pressao(pressao):
    sistolica, diastolica = map(int, pressao.split('/'))
    if sistolica < 120 and diastolica < 80:
        return 'Normal'
    elif (120 <= sistolica <= 129) and (80 <= diastolica <= 84):
        return 'Pré-hipertensão'
    elif (130 <= sistolica <= 139) or (85 <= diastolica <= 89):
        return 'Hipertensão Estágio 1'
    else:
        return 'Hipertensão Estágio 2'

df['Categoria Pressão'] = df['Pressão Sanguínea'].apply(categorizar_pressao)
contagem_categoria = df.groupby('Gênero')['Categoria Pressão'].value_counts().unstack()

print(contagem_categoria)
```

```python
#10. Pessoas com frequências cardíacas acima de 70 dão mais passos que pessoas com frequência cardíaca menor ou igual a 70? (Use a média)
acima_70 = df[df['Frequência cardíaca'] > 70]
menor_igual_70 = df[df['Frequência cardíaca'] <= 70]
media_acima_70 = acima_70['Passos diários'].mean()
media_menor_igual_70 = menor_igual_70['Passos diários'].mean()

print(f"Média de passos para frequência cardíaca acima de 70: {media_acima_70:.2f}")
print(f"Média de passos para frequência cardíaca menor ou igual a 70: {media_menor_igual_70:.2f}")


```
